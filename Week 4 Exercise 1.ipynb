The equation
$$z(x,y) = z_0\sin\Big(k\sqrt{x^2+y^2}-\omega t\Big)$$
describes a circular set of waves centered on the origin that propagate outwards as time progresses. Your task is to create an animated representation of this wave.

For simplicity, start with $z_0=1$, $k=1$, and let $\phi=\omega t$. Depending on how you set things up, you may want to adjust $z_0$ and $k$ to make your animation look good. The advantage of setting $\phi=\omega t$ is that we can run our animation from $\phi=0$ to $\phi=2\pi$ to end up with an animation that loops smoothly.

First, create a static image of the wave for $\phi=0$. It should resemble a bullseye.

```python
import matplotlib.pyplot as plt
import numpy as np

z0 = 1
k = 1
phi = 0

L = 500
d = 35

grid = np.empty([L,L])
x = np.linspace(-d/2,d/2,L)
y = np.linspace(-d/2,d/2,L)
xx,yy = np.meshgrid(x,y)

z = z0*np.sin(k*np.sqrt(xx**2 + yy**2) - phi)
        
plt.imshow(z, cmap = "jet")
plt.axis("off")
plt.show()
```

Now modify your program to create an animation of the wave with $\phi$ taking 100 steps (one step per frame) starting at 0 and ending just before 2$\pi$ (so that the animation loops smoothly).

```python
plt.close("all")

fig = plt.figure()
num_steps = 100
delay_in_ms = 20

z0 = 1
k = 1

L = 500
d = 30
phi = 0

x = np.linspace(-d/2,d/2,L)
y = np.linspace(-d/2,d/2,L)
xx,yy = np.meshgrid(x,y)

z = z0*np.sin(k*np.sqrt(xx**2 + yy**2) - phi)

pic = plt.imshow(grid, cmap = "jet")
plt.axis("off")

def animate(frame):
    phi = 2*pi*frame/100
    z = z0*np.sin(k*np.sqrt(xx**2 + yy**2) - phi)
    
    pic.set_data(z)
    
```

```python
#show your animation
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

#you will likely need to edit and uncomment the next line:
#plt.rcParams["animation.ffmpeg_path"] = r"full/path/to/ffmpeg_binary"

anim = FuncAnimation(fig, animate, frames = 100, interval = 20)
HTML(anim.to_html5_video())
```
